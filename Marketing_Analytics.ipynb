```python
df.columns = df.columns.str.replace(' ', '')
df.info()
```

```python
df['Income'] = df['Income'].str.replace('$', '')
df['Income'] = df['Income'].str.replace(',', '')
df['Income'] = df['Income'].astype('float')
```

```python
df.info()
```

```python
sns.boxplot(y='Income', data = df)
```

```python
df["Income"] = df["Income"].fillna(df['Income'].median())
```

```python
df['Income'].isnull().sum()
```

```python
df_plot = df.drop(['ID', 'Education', 'Marital_Status', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 'Complain'], axis=1)
```

```python
columns = ['Year_Birth', 'Income', 'Kidhome', 'Teenhome', 'Recency',
       'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
       'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
       'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
       'NumWebVisitsMonth']
sns.set_palette("Set1")
fig, ax = plt.subplots(4, 4,figsize = (16, 10))
ax = ax.flatten()
for i,col in enumerate(columns):
    sns.boxplot(x = df_plot[col], ax = ax[i])
    
plt.suptitle("Outiers Analysis")
fig.tight_layout()
```

```python
df = df[df['Year_Birth'] > 1900].reset_index(drop = True)
df.head()
```

```python
countplot = ['Education', 'Marital_Status', 'Kidhome', 'Teenhome', 'Complain', 'Response', 'Country', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
fig, ax = plt.subplots(int(len(countplot) / 2), 2, figsize = (15, 20))
plt.suptitle('Unique Value Count Plot', y = 1.0, fontsize = 25)

ax = ax.flatten()
for i, c in enumerate(countplot):
    sns.countplot(x = c, data = df, ax = ax[i])
fig.tight_layout()
```

```python
cam_success = pd.DataFrame(round(df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].mean()*100, 2), 
                           columns=['Percent']).reset_index()
```

```python
cam_success
```

```python
sns.catplot(y = 'Percent',x = "index", data = cam_success, color = "green")
plt.xticks(rotation=90)
```

```python
spending = pd.DataFrame(round(df[['MntWines',
       'MntFruits', 'MntMeatProducts', 'MntFishProducts',
       'MntSweetProducts', 'MntGoldProds']].mean(), 1), columns=['Average']).sort_values(by='Average').reset_index()

# plot
ax = sns.barplot(x='Average', y='index', data=spending, palette='Blues')
plt.ylabel('Amount spent on...')
```
